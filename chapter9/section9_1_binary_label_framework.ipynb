# 2値ラベルの確率整理（Ω・E・更新式）

この章は、**2値ラベル問題**を
- Ω（結果空間）
- E（イベント＝部分集合）
- 更新式（逐次観測）
で整理し、**PBTにもCBTにも使える最短形**を作る。

測度論は使わない。集合と条件付き確率の分解だけで完結させる。


## 1. 登場人物の整理（Ωの導入）

2値ラベル（真の状態）と観測を**同じ層の要素**として扱う。

- 真の状態：
  \( D\in\{1,0\} \)
- 観測：
  \( R_1, R_2, \dots \in\{1,0\} \)

観測を\(i\)回まで扱うなら


\[
\Omega_i=\{(d,r_1,\dots,r_i)\mid d\in\{1,0\},\ r_k\in\{1,0\}\}
\]

ここで **Ωは「登場人物を整列するためのスイッチ」**として使う。


## 2. イベントの定義（Eの差し替え）

観測結果をイベントとして定義する：

\[
E_k=\{R_k=r_k\}
\]

\(i\)回分をまとめたイベント：

\[
E_{1:i}=igcap_{k=1}^{i}E_k
\]

つまり

\[
E_{1:i}=E_{1:i-1}\cap E_i
\]

**差し替えるのはこの \(E\) だけ**。ここが最短の分岐点。


## 3. 全確率（分母は「分割の足し算」）

二値なので必ず

\[
E_{1:i}=(E_{1:i}\cap\{D=1\})\cup(E_{1:i}\cap\{D=0\})
\]

互いに素なので

\[
P(E_{1:i})=P(E_{1:i}\cap\{D=1\})+P(E_{1:i}\cap\{D=0\})
\]

よって

\[
P(E_{1:i})=P(E_{1:i}\mid D=1)P(D=1)+P(E_{1:i}\mid D=0)P(D=0)
\]

**分母は「分割して足す」**。ここで条件付けに戻さない。


## 4. 条件付きの分解（\(\cap\) を右側に押し込む）

基本恒等式：

\[
P(A\cap B\mid C)=P(A\mid B\cap C)P(B\mid C)
\]

これを\(A=E_i\), \(B=E_{1:i-1}\), \(C=\{D=d\}\) に当てると

\[
P(E_{1:i}\mid D=d)=P(E_i\mid E_{1:i-1}\cap\{D=d\})\,P(E_{1:i-1}\mid D=d)
\]

**新しい条件を右側に積む**感覚をここで固定する。


## 5. 尤度 \(L_i\) と更新式 \(q_i\)

尤度（当てはまり）を部品化する：

\[
L_i(d)=P(E_{1:i}\mid D=d)
\]

すると更新は

\[
L_i(d)=P(E_i\mid E_{1:i-1}\cap\{D=d\})\,L_{i-1}(d)
\]

事後確率を

\[
q_i=P(D=1\mid E_{1:i})
\]

と置けば

\[
q_i=rac{L_i(1)P(D=1)}{L_i(1)P(D=1)+L_i(0)P(D=0)}
\]

**分母を作ると分子が確定する**ので、CBTではここを最短化する。


## 6. CBT最短メモ（自分用の符号化）

CBTでは説明不要なので、紙に吐くのはこれだけ：

- (Ω) 登場人物の整列
- (E) 差し替えイベント
- (A)(B) 分子と反対側の重み
- (Ans) \(A/(A+B)\) か \(1/(1+m)\)

二値なら

\[
m=rac{P(E\mid D=0)P(D=0)}{P(E\mid D=1)P(D=1)},\quad	ext{Ans}=rac{1}{1+m}
\]

を固定すれば十分。


## 7. まとめ（最短形）

- Ωは登場人物の整列
- Eは差し替え点
- 分母は分割の足し算
- \(\cap\) を条件側に押し込む
- 最後は \(A/(A+B)\) or \(1/(1+m)\)

この型が固定できれば、検査問題・逐次観測・判定問題まで同型で処理できる。
