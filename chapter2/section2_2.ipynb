```python
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

boston = load_boston()
#print(boston.DESCR)

#1.必要なデータの読み込み
df=pd.DataFrame(boston.data,columns = boston.feature_names) #説明変数を読込
df['MEDV'] = boston.target #目的変数を読込
x=df.RM.to_frame()
y=df.MEDV
#print(x)
#print(y)

from sklearn.model_selection import train_test_split
#2.訓練データと評価データに分割
(x_train, x_test, y_train, y_test) = train_test_split(x, y, test_size=0.3)

from sklearn.linear_model import LinearRegression
#3.モデル選択 -> 線形回帰
lr = LinearRegression()

#4.学習
lr.fit(x_train, y_train)

#5. 評価データを使った予測
y_pred=lr.predict(x_test)
print(y_pred[0]*1000)

pred_table = pd.DataFrame({"y_true": y_test, "y_pred": y_pred})
pred_table["residual"] = pred_table["y_true"] - pred_table["y_pred"]
display(pred_table.head(10))

plt.figure()
plt.scatter(pred_table["y_true"], pred_table["y_pred"], alpha=0.7)
plt.plot([pred_table["y_true"].min(), pred_table["y_true"].max()],
         [pred_table["y_true"].min(), pred_table["y_true"].max()],
         color="#E45756", linestyle="--")
plt.xlabel("実測値")
plt.ylabel("予測値")
plt.title("実測値と予測値の比較")
plt.tight_layout()
plt.show()

from sklearn.metrics import mean_squared_error
#6.モデルの評価
mse=mean_squared_error(y_test, y_pred)
print(mse)

plt.figure()
plt.hist(pred_table["residual"], bins=20, color="#72B7B2", edgecolor="white")
plt.title("残差の分布")
plt.xlabel("残差")
plt.ylabel("度数")
plt.tight_layout()
plt.show()
```
