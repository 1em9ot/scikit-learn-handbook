```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

#1.必要なデータの読み込み
npArray = np.loadtxt("in.csv", delimiter = ",", dtype = "float",skiprows=1)
x = npArray[:, 1:3]
y = npArray[:, 3:4].ravel()
#print("説明変数を出力\n", x)
#print("目的変数を出力\n", y)

from sklearn.model_selection import train_test_split
#2.訓練データと評価データに分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
#print("訓練用の説明変数を出力\n", x_train)
#print("訓練用の目的変数を出力\n", y_train)
#print("評価用の説明変数を出力\n", x_test)
#print("評価用の目的変数を出力\n", y_test)

from sklearn import tree
#3.モデル選択 -> 決定木
clf = tree.DecisionTreeClassifier()

#4.学習
clf.fit(x_train, y_train)

#5.予測
predict = clf.predict(x_test)
print("評価結果を出力\n", predict)

result_df = pd.DataFrame({"y_true": y_test, "y_pred": predict})
display(result_df.head(10))
display(result_df["y_pred"].value_counts().rename_axis("label").to_frame("count"))

from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
#6. モデルの評価
accuracy = accuracy_score(y_test, predict)
print("正解率を出力\n", accuracy)

ConfusionMatrixDisplay.from_predictions(y_test, predict, cmap="Blues")
plt.title("混同行列")
plt.tight_layout()
plt.show()

plt.figure()
plt.bar(["accuracy"], [accuracy], color="#4C78A8")
plt.ylim(0, 1)
plt.ylabel("score")
plt.title("正解率")
plt.tight_layout()
plt.show()
```
